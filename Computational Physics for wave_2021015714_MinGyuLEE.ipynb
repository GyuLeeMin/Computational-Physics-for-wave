```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_wavefunction(x, A, mu, sigma):
    """
    Generates a wave function using a Gaussian function.

    Parameters:
    x (numpy array): Array of values for the independent variable.
    A (float): Amplitude (height) of the Gaussian function.
    mu (float): Mean of the Gaussian function.
    sigma (float): Standard deviation of the Gaussian function.

    Returns:
    numpy array: Array of values for the Gaussian wave function.
    """
    psi = A * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
    return psi

# Set the parameters for wave function generation
A = 1.0  # Amplitude
mu = 0.0  # Mean
sigma = 1.0  # Standard deviation

# Set the range for x
x = np.linspace(-5, 5, 100)

# Generate the Gaussian wave function
psi = gaussian_wavefunction(x, A, mu, sigma)

# Plot the results
plt.plot(x, psi)
plt.xlabel('x')
plt.ylabel('Wavefunction')
plt.title('Gaussian Wavefunction')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating the wave function
def wave_function(x, mean, std_dev):
    return np.exp(-(x - mean)**2 / (2 * std_dev**2)) * np.cos(2 * np.pi * x)

# Setting the parameters
mean = 0  # Mean
std_dev = 1  # Standard deviation
num_points = 1000  # Number of data points to sample

# Sampling from the Gaussian distribution
x = np.linspace(-10, 10, num_points)
samples = np.random.normal(mean, std_dev, num_points)

# Calculating the wave function values
y = wave_function(x, mean, std_dev)

# Plotting the graph
plt.plot(x, y, label='Wave Function')
plt.scatter(samples, np.zeros_like(samples), color='red', label='Random Samples')
plt.xlabel('x')
plt.ylabel('Wave Function Value')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generating the wave function
def wave_function(x, mean, std_dev):
    return np.exp(-(x - mean)**2 / (2 * std_dev**2)) * np.cos(2 * np.pi * x)

# Setting the parameters
mean = 0  # Mean
std_dev = 1  # Standard deviation
num_points = 1000  # Number of data points to sample

# Sampling from the Gaussian distribution
x = np.linspace(-10, 10, num_points)
samples = np.random.normal(mean, std_dev, num_points)

# Calculating the wave function values
y = wave_function(x, mean, std_dev)

# Calculating the probability density function (PDF)
pdf = norm.pdf(x, mean, std_dev)

# Plotting the graph
plt.plot(x, y, label='Wave Function')
plt.plot(x, pdf, label='Probability Density Function (PDF)')
plt.scatter(samples, np.zeros_like(samples), color='red', label='Random Samples')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.show()
```
